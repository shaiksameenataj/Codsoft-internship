```python
import tkinter as tk
from tkinter import ttk
import random
import string

class PasswordGenerator:
    def __init__(self, root):
        self.root = root
        self.root.title("Password Generator")
        self.root.geometry("800x600")  # Set the window size to 800x600
        self.root.configure(bg="#2b2b2b")  # Set the background color to dark gray

        # Create a label and combobox for password strength
        self.strength_label = tk.Label(root, text="Password Strength:", bg="#2b2b2b", fg="#ffffff", font=("Arial", 14))
        self.strength_label.grid(row=0, column=0, padx=10, pady=10)

        self.strength_combo = ttk.Combobox(root, values=["Strong", "Moderate", "Weak"], font=("Arial", 14))
        self.strength_combo.current(0)  # Set default value to "Strong"
        self.strength_combo.grid(row=0, column=1, padx=10, pady=10)

        # Create a label and entry for password length
        self.length_label = tk.Label(root, text="Password Length:", bg="#2b2b2b", fg="#ffffff", font=("Arial", 14))
        self.length_label.grid(row=1, column=0, padx=10, pady=10)

        self.length_entry = tk.Entry(root, width=10, font=("Arial", 14), bg="#4b4b4b", fg="#ffffff")
        self.length_entry.grid(row=1, column=1, padx=10, pady=10)

        # Create a button to generate password
        self.generate_button = tk.Button(root, text="Generate Password", command=self.generate_password, bg="#4CAF50", fg="#ffffff", font=("Arial", 14))
        self.generate_button.grid(row=2, column=0, columnspan=2, padx=10, pady=10)

        # Create a label to display generated password
        self.password_label = tk.Label(root, text="", bg="#2b2b2b", fg="#ffffff", font=("Arial", 14))
        self.password_label.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

    def generate_password(self):
        strength = self.strength_combo.get()
        try:
            length = int(self.length_entry.get())
            if length < 8:
                self.password_label.config(text="Password length should be at least 8 characters.")
                return
        except ValueError:
            self.password_label.config(text="Invalid input. Please enter a positive integer.")
            return

        characters = string.ascii_letters + string.digits + string.punctuation
        if strength == "Weak":
            characters = string.ascii_letters
        elif strength == "Moderate":
            characters = string.ascii_letters + string.digits

        password = ''.join(random.choice(characters) for _ in range(length))
        self.password_label.config(text=f"Generated Password: {password}")

if __name__ == "__main__":
    root = tk.Tk()
    app = PasswordGenerator(root)
    root.state("zoomed")  # Open the window in full screen
    root.mainloop()
```
