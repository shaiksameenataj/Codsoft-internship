```python
import tkinter as tk
from math import sqrt, sin, cos, tan, log10

def calculate():
    try:
        result = eval(entry.get())
        label_result.config(text=f"Result: {result}", fg="green")
    except Exception as e:
        label_result.config(text=f"Error: {str(e)}", fg="red")

def clear():
    entry.delete(0, tk.END)
    label_result.config(text="Result: ", fg="black")

root = tk.Tk()
root.title("Calculator")
root.configure(background="black")  # light gray background

entry = tk.Entry(root, width=35, borderwidth=5, bg="white", fg="black")
entry.grid(row=0, column=0, columnspan=4)

button_7 = tk.Button(root, text="7", padx=40, pady=20, command=lambda: entry.insert(tk.END, "7"), bg="white", fg="black")
button_8 = tk.Button(root, text="8", padx=40, pady=20, command=lambda: entry.insert(tk.END, "8"), bg="white", fg="black")
button_9 = tk.Button(root, text="9", padx=40, pady=20, command=lambda: entry.insert(tk.END, "9"), bg="white", fg="black")
button_div = tk.Button(root, text="/", padx=40, pady=20, command=lambda: entry.insert(tk.END, "/"), bg="grey", fg="black")

button_4 = tk.Button(root, text="4", padx=40, pady=20, command=lambda: entry.insert(tk.END, "4"), bg="white", fg="black")
button_5 = tk.Button(root, text="5", padx=40, pady=20, command=lambda: entry.insert(tk.END, "5"), bg="white", fg="black")
button_6 = tk.Button(root, text="6", padx=40, pady=20, command=lambda: entry.insert(tk.END, "6"), bg="white", fg="black")
button_mul = tk.Button(root, text="*", padx=40, pady=20, command=lambda: entry.insert(tk.END, "*"), bg="grey", fg="black")

button_1 = tk.Button(root, text="1", padx=40, pady=20, command=lambda: entry.insert(tk.END, "1"), bg="white", fg="black")
button_2 = tk.Button(root, text="2", padx=40, pady=20, command=lambda: entry.insert(tk.END, "2"), bg="white", fg="black")
button_3 = tk.Button(root, text="3", padx=40, pady=20, command=lambda: entry.insert(tk.END, "3"), bg="white", fg="black")
button_sub = tk.Button(root, text="-", padx=40, pady=20, command=lambda: entry.insert(tk.END, "-"), bg="grey", fg="black")

button_0 = tk.Button(root, text="0", padx=40, pady=20, command=lambda: entry.insert(tk.END, "0"), bg="white", fg="black")
button_dec = tk.Button(root, text=".", padx=40, pady=20, command=lambda: entry.insert(tk.END, "."), bg="white", fg="black")
button_add = tk.Button(root, text="+", padx=40, pady=20, command=lambda: entry.insert(tk.END, "+"), bg="grey", fg="black")
button_eq = tk.Button(root, text="=", padx=40, pady=20, command=calculate, bg="grey", fg="black")

button_sin = tk.Button(root, text="sin", padx=35, pady=20, command=lambda: entry.insert(tk.END, "sin("), bg="grey", fg="black")
button_cos = tk.Button(root, text="cos", padx=35, pady=20, command=lambda: entry.insert(tk.END, "cos("), bg="grey", fg="black")
button_tan = tk.Button(root, text="tan", padx=35, pady=20, command=lambda: entry.insert(tk.END, "tan("), bg="grey", fg="black")
button_sqrt = tk.Button(root, text="âˆš", padx=40, pady=20, command=lambda: entry.insert(tk.END, "sqrt("), bg="grey", fg="black")
button_log = tk.Button(root, text="log", padx=35, pady=20, command=lambda: entry.insert(tk.END, "log10("), bg="grey", fg="black")
button_clear = tk.Button(root, text="Clear", padx=30, pady=20, command=clear, bg="red", fg="white")

button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)
button_div.grid(row=1, column=3)

button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)
button_mul.grid(row=2, column=3)

button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)
button_sub.grid(row=3, column=3)

button_0.grid(row=4, column=0)
button_dec.grid(row=4, column=1)
button_add.grid(row=4, column=2)
button_eq.grid(row=4, column=3)

button_sin.grid(row=5, column=0)
button_cos.grid(row=5, column=1)
button_tan.grid(row=5, column=2)
button_sqrt.grid(row=5, column=3)

button_log.grid(row=6, column=0)
button_clear.grid(row=6, column=1, columnspan=3)

label_result = tk.Label(root, text="Result: ", bg="lightgreen", fg="black")
label_result.grid(row=7, column=0, columnspan=4)

root.mainloop()
```
