```python
import tkinter as tk
from random import choice

class RockPaperScissors:
    def __init__(self):
        self.root = tk.Tk()
        self.root.title("Rock Paper Scissors")
        self.root.geometry("300x200")
        self.user_score = 0
        self.computer_score = 0

        self.user_score_label = tk.Label(self.root, text="User Score: 0", font=("Arial", 12), fg="blue")
        self.user_score_label.pack()

        self.computer_score_label = tk.Label(self.root, text="Computer Score: 0", font=("Arial", 12), fg="red")
        self.computer_score_label.pack()

        self.result_label = tk.Label(self.root, text="", font=("Arial", 12), fg="green")
        self.result_label.pack()

        self.rock_button = tk.Button(self.root, text="Rock", command=self.rock, font=("Arial", 12), bg="gray")
        self.rock_button.pack()

        self.paper_button = tk.Button(self.root, text="Paper", command=self.paper, font=("Arial", 12), bg="grey")
        self.paper_button.pack()

        self.scissors_button = tk.Button(self.root, text="Scissors", command=self.scissors, font=("Arial", 12), bg="grey")
        self.scissors_button.pack()

        self.play_again_button = tk.Button(self.root, text="Play Again", command=self.play_again, font=("Arial", 12), bg="orange")
        self.play_again_button.pack()

    def computer_choice(self):
        return choice(["rock", "paper", "scissors"])

    def rock(self):
        computer = self.computer_choice()
        if computer == "rock":
            self.result_label.config(text="It's a tie!", fg="yellow")
        elif computer == "paper":
            self.computer_score += 1
            self.computer_score_label.config(text=f"Computer Score: {self.computer_score}", fg="red")
            self.result_label.config(text="Paper covers rock! Computer wins!", fg="red")
        else:
            self.user_score += 1
            self.user_score_label.config(text=f"User Score: {self.user_score}", fg="blue")
            self.result_label.config(text="Rock smashes scissors! User wins!", fg="blue")

    def paper(self):
        computer = self.computer_choice()
        if computer == "paper":
            self.result_label.config(text="It's a tie!", fg="yellow")
        elif computer == "rock":
            self.user_score += 1
            self.user_score_label.config(text=f"User Score: {self.user_score}", fg="blue")
            self.result_label.config(text="Paper covers rock! User wins!", fg="blue")
        else:
            self.computer_score += 1
            self.computer_score_label.config(text=f"Computer Score: {self.computer_score}", fg="red")
            self.result_label.config(text="Scissors cut paper! Computer wins!", fg="red")

    def scissors(self):
        computer = self.computer_choice()
        if computer == "scissors":
            self.result_label.config(text="It's a tie!", fg="yellow")
        elif computer == "paper":
            self.user_score += 1
            self.user_score_label.config(text=f"User Score: {self.user_score}", fg="blue")
            self.result_label.config(text="Scissors cut paper! User wins!", fg="blue")
        else:
            self.computer_score += 1
            self.computer_score_label.config(text=f"Computer Score: {self.computer_score}", fg="red")
            self.result_label.config(text="Rock smashes scissors! Computer wins!", fg="red")

    def play_again(self):
        self.user_score = 0
        self.computer_score = 0
        self.user_score_label.config(text="User Score: 0", fg="blue")
        self.computer_score_label.config(text="Computer Score: 0", fg="red")
        self.result_label.config(text="", fg="green")

    def run(self):
        self.root.mainloop()

if __name__ == "__main__":
    game = RockPaperScissors()
    game.run()
```
